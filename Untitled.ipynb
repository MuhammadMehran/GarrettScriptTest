```python
import sqlite3
from bs4 import BeautifulSoup
from requests import get
import time
```

```python
usr_agent = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) '
                      'Chrome/61.0.3163.100 Safari/537.36'}




urls = []
with open('urls.txt') as f:
    for line in f:
        url = line.strip()
        if not line.startswith('http'):
            url = 'http://' + line
        urls.append(url)         


```

```python
q = ''

for url in urls[:10]:
    q += 'site:'+url + ' OR '
q = q[:-3]
q = q + 'fat'
q
```

```python
len(q)
```

```python

escaped_search_term = q.replace(' ', '+')
number_results = 100
language_code = 'en'

# google_url = 'https://www.google.com/search?q={}&num={}&hl={}'.format(escaped_search_term, number_results+1,
#                                                                     language_code)


google_url = 'https://www.google.com/search?q={}&num={}&hl={}+start={}'.format(escaped_search_term, number_results+1,
                                                                    language_code, 300)

response = get(google_url, headers=usr_agent)    
response.raise_for_status()

raw_html = response.text


```

```python
results = []
json_return = []
soup = BeautifulSoup(raw_html, 'html.parser')
result_block = soup.find_all('div', attrs={'class': 'g'})
for result in result_block:
    try:
        link = result.find('a', href=True)
        title = result.find('h3')
        content = 'Not Available'
        content = result.find('div', attrs={'class': 'IsZvec'}).text
#         spans = result.findAll('span')
#         for span in spans:
#             if span.get('class') == None and span.get('style') == None:
#                 text = span.text.strip()
#                 if text != '':
#                     content = text
#                     break
        if link and title and content:
            # print(link['href'], title.text, content)
            # print()
            json_return.append({'title': title.text, 'link': link['href'], 'content': content})
            results.append([title.text, link['href'], content])
    except:
        pass
    
```
